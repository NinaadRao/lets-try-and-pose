Install required dependences for the detectron2 repo and run apply_net.py to get the required densepose images

```python
%%shell
eval "$(conda shell.bash hook)"
conda activate pws
pip install av
```

```python
%%shell
eval "$(conda shell.bash hook)"
conda activate pws
cp pose-with-style/DATASET/DeepFashion/train/* detectron2/projects/DensePose/images/train/
cd detectron2/projects/DensePose
python apply_net.py dump configs/densepose_rcnn_R_101_FPN_DL_s1x.yaml https://dl.fbaipublicfiles.com/densepose/densepose_rcnn_R_101_FPN_DL_s1x/165712116/model_final_844d15.pkl images/train --output train.pkl -v
```

```python
%%shell
eval "$(conda shell.bash hook)"
conda activate pws
cp pose-with-style/DATASET/DeepFashion/test/* detectron2/projects/DensePose/images/test/
cd detectron2/projects/DensePose
python apply_net.py dump configs/densepose_rcnn_R_101_FPN_DL_s1x.yaml https://dl.fbaipublicfiles.com/densepose/densepose_rcnn_R_101_FPN_DL_s1x/165712116/model_final_844d15.pkl images/test --output test.pkl -v
```

Now, convert the generated pickle files from apply_net to .png images using pickle2png.py

```python
%%shell
eval "$(conda shell.bash hook)"
conda activate pws
cp pose-with-style/util/pickle2png.py detectron2/projects/DensePose
```

```python
%%shell
eval "$(conda shell.bash hook)"
conda activate pws
cd detectron2/projects/DensePose
python pickle2png.py --pickle_file train.pkl --save_path ../../../pose-with-style/DATASET/densepose/train
```

```python
%%shell
eval "$(conda shell.bash hook)"
conda activate pws
cd detectron2/projects/DensePose
python pickle2png.py --pickle_file test.pkl --save_path ../../../pose-with-style/DATASET/densepose/test
```
